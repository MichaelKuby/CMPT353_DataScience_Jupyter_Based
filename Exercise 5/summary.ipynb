```python
import numpy as np
import pandas as pd
from scipy import stats
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
### advice on how to iterate over various csv files in the same folder found here:
### https://swcarpentry.github.io/python-novice-gapminder/14-looping-data-sets/
### Note that you will need to set the folder to get the code to work

i = 1
folder = '/Users/MichaelKuby/Documents/GitHub/CMPT353_Computational_Data_Science/CMPT318_DataScience_Jupyter_Based/Exercise 5'
for filename in Path(folder).glob("*.csv"):
    x = pd.read_csv(filename)
    
    ## compute the mean, standard deviation, and range of both x and y ##
    print ("DataSet", i)
    print(x.describe())
    print('')
    ## compute the correlation coefficient (r) between the x and y variables
    
    print("Correlational Coefficient for x and y of dataset", i, ":", stats.linregress(x['x'], x['y']).rvalue)
    print('')
    
    
    i += 1
```
